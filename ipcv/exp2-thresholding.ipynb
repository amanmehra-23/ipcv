```python
L = 256
```

```python
def apply_image_negative(image_array,L):
    negative_array = (L-1)-image_array
    return negative_array
```

```python
negative_array = apply_image_negative(array1,L)
negative_image = Image.fromarray(negative_array)
plt.imshow(negative_image,cmap='gray')
plt.title("Negative Image")
```

```python
def apply_thresholding(image_array,threshold_value):
    threshold_array = (image_array > threshold_value).astype(np.uint8)
    return threshold_array
```

```python
threshold_array = apply_thresholding(array1,127)
threshold_image = Image.fromarray(threshold_array)
plt.imshow(threshold_array,cmap='gray')
plt.title("Thresholding")
```

```python
def grey_level_without_bg(image_array,lower_bound,upper_bound):
    sliced_image = np.where((image_array>= lower_bound) & (image_array<= upper_bound),255,0).astype(np.uint8)
    return sliced_image
```

```python
grey_without_bg = grey_level_without_bg(array1,110,220)
plt.imshow(grey_without_bg,cmap='gray')
plt.title("Grey Level Slicing Without Background")
    
```

```python
grey_with_bg = grey_level_with_bg(array1,170,200)
plt.imshow(grey_with_bg,cmap='gray')
plt.title("Grey Level Slicing With Background")
```
