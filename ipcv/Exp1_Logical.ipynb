```python
import numpy as np 
import matplotlib.pyplot as plt
from PIL import Image
import pandas as pd
```

```python
data = pd.read_csv("mnist_train.csv")
```

```python
rownum1= 10
rownum2 =20
if -1 < rownum1 <len(data):
    pixels1 = data.iloc[rownum1,1:].values.reshape(28,28)
    array1 = np.array(pixels1,dtype=np.uint8)

    if -1< rownum2 < len(data):
        pixels2 = data.iloc[rownum2,1:].values.reshape(28,28)
        array2 = np.array(pixels2,dtype=np.uint8)
```

```python
image1 = Image.fromarray(array1)
plt.imshow(image1,cmap='gray')
```

```python
image2 = Image.fromarray(array2)
plt.imshow(image2,cmap='gray')
```

```python
addition_array = np.clip(array1+array2,0,255).astype(np.uint8)
```

```python
addition_img = Image.fromarray(addition_array)
plt.imshow(addition_img,cmap='gray')
plt.title("Addition Image")
plt.axis('off')
```

```python
subtraction_array = np.clip(array1-array2,0,255).astype(np.uint8)
subtraction_img = Image.fromarray(subtraction_array)
plt.imshow(subtraction_img,cmap='gray')
plt.title("subtracted image")
plt.axis('off')
```

```python
x,y,w,h = 10,10,40, 100
cropped_image = image1.crop((x,y,x+w,y+h))
plt.imshow(cropped_image,cmap='gray')
plt.title("Cropped Image")
```

```python
image1_binary = (array1 > 127).astype(np.int8)
image2_binary = (array2 > 127).astype(np.uint8)
and_array = np.logical_and(image1_binary,image2_binary).astype(np.uint8)
and_image = Image.fromarray(and_array)
plt.title("AND Image")
plt.imshow(and_image,cmap='gray')
```

```python
image1_binary = (array1 > 127).astype(np.uint8)
image2_binary = (array2 > 127).astype(np.uint8)
or_array = np.logical_or(image1_binary,image2_binary).astype(np.uint8)
or_image = Image.fromarray(or_array)
plt.title("OR Image")
plt.imshow(or_image,cmap='gray')
```

```python
image1_binary = (array1 > 127).astype(np.uint8)
not_array = np.logical_not(image1_binary)
not_image = Image.fromarray(not_array)
plt.title("NOT Image")
plt.imshow(not_image,cmap='gray')
```

```python
image1_binary =  (array1 > 127).astype(np.uint8)
image2_binary = (array2 > 127).astype(np.uint8)
xor_array = np.logical_xor(image1_binary,image2_binary).astype(np.uint8)
xor_image = Image.fromarray(xor_array)
plt.title("XOR Image")
plt.imshow(xor_image,cmap='gray')
```
