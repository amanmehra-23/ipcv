```python
import pandas as pd
import numpy as np
from PIL import Image
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("mnist_train.csv")
```

```python
rownum1 = 10
if -1 < rownum1 <= len(data):
    pixels = data.iloc[rownum1,1:].values.reshape(28,28)
    array1 = np.array(pixels)
```

```python
plt.imshow(array1,cmap='gray')
plt.title("Original Image")
```

```python
def histogram_equalisation(image_array,L):
    histogram,_ = np.histogram(image_array.flatten(),bins=np.arange(L+1))
    rows,cols = image_array.shape
    total_pixels = rows * cols
    pmf = histogram /total_pixels
    cdf = np.cumsum(pmf)
    new_grey_levels = (cdf * (L-1)).astype(np.uint8)
    equalised_array = new_grey_levels[image_array]
    return equalised_array
```

```python
histogram_image = histogram_equalisation(array1,256)
plt.imshow(histogram_image,cmap='gray')
plt.title("Histogram Equalised Image")
```
